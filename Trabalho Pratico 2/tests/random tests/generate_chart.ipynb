```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
```

```python
data = pd.read_csv("tests.csv", sep=",", header=None, index_col=0)
```

```python
data
```

```python
x = data.index.values
y = data.mean(axis = 1).values
error = data.std(axis = 1).values
error = 2.6*error

#x = np.squeeze(np.array(x))
#y = np.squeeze(np.array(y))
#error = np.squeeze(np.array(error))
```

```python
rcParams['figure.figsize'] = 10, 5

plt.scatter(x[0::3],y[0::3])
plt.errorbar(x[0::3], y[0::3], yerr=error[0::3], uplims=False, lolims=False)
plt.ylim(bottom=-0.005)
plt.title("Análise Estatística de Desempenho com Programação Dinâmica (Média e Desvio Padrão)")
plt.ylabel("Tempo de execução")
plt.xlabel("Número de Diamantes")
plt.show()
```

```python
data2 = pd.read_csv("times_brute_force.csv", sep=",", header=None)
```

```python
data2
```

```python
x_2 = data2.groupby(0).mean().index
y_2 = data2.groupby(0).mean()
error_2 = data2.groupby(0).std()

x_2 = np.squeeze(np.array(x_2))
y_2 = np.squeeze(np.array(y_2))
error_2 = np.squeeze(np.array(error_2))
```

```python
rcParams['figure.figsize'] = 8, 4.5

plt.errorbar(x, y, yerr=error, uplims=True, lolims=True, label='Programação dinâmica')
plt.scatter(x, y)

plt.errorbar(x_2, y_2, yerr=error_2, uplims=True, lolims=True, label='Força bruta')
plt.scatter(x_2, y_2)

plt.legend(loc='upper right')

plt.title("Análise Prática de Desempenho (programação dinâmica X força bruta)")
plt.ylabel("Tempo de execução (s)")
plt.xlabel("Tamanho da entrada (número de diamantes)")
plt.show()
```
