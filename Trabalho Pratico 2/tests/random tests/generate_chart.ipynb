```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
```

```python
data = pd.read_csv("tests.csv", sep=",", header=None, index_col=0)
```

```python
data
```

```python
x = data.index.values
y = data.mean(axis = 1).values
error = data.std(axis = 1).values
error = 2.6*error

#x = np.squeeze(np.array(x))
#y = np.squeeze(np.array(y))
#error = np.squeeze(np.array(error))
```

```python
rcParams['figure.figsize'] = 10, 5

plt.scatter(x[0::3],y[0::3])
plt.errorbar(x[0::3], y[0::3], yerr=error[0::3], uplims=False, lolims=False)
plt.ylim(bottom=-0.005)
plt.title("AnÃ¡lise EstatÃ­stica de Desempenho com ProgramaÃ§Ã£o DinÃ¢mica (MÃ©dia e Desvio PadrÃ£o)")
plt.ylabel("Tempo de execuÃ§Ã£o")
plt.xlabel("NÃºmero de Diamantes")
plt.show()
```

```python
data2 = pd.read_csv("times_brute_force.csv", sep=",", header=None)
```

```python
data2
```

```python
x_2 = data2.groupby(0).mean().index
y_2 = data2.groupby(0).mean()
error_2 = data2.groupby(0).std()

x_2 = np.squeeze(np.array(x_2))
y_2 = np.squeeze(np.array(y_2))
error_2 = np.squeeze(np.array(error_2))
```

```python
rcParams['figure.figsize'] = 8, 4.5

plt.errorbar(x, y, yerr=error, uplims=True, lolims=True, label='ProgramaÃ§Ã£o dinÃ¢mica')
plt.scatter(x, y)

plt.errorbar(x_2, y_2, yerr=error_2, uplims=True, lolims=True, label='ForÃ§a bruta')
plt.scatter(x_2, y_2)

plt.legend(loc='upper right')

plt.title("AnÃ¡lise PrÃ¡tica de Desempenho (programaÃ§Ã£o dinÃ¢mica X forÃ§a bruta)")
plt.ylabel("Tempo de execuÃ§Ã£o (s)")
plt.xlabel("Tamanho da entrada (nÃºmero de diamantes)")
plt.show()
```
